```python
import pandas as pd 
import numpy as np

biallelic_Dlabrax_maf=pd.read_csv(r"maf_file.csv", header=None, sep=" ")
biallelic_Dlabrax_maf
```

```python
# prepare the .genotype file for the BayPass analysis

# ==============================

def BayPass_genotyping_file(infile, colnames):
    TOT_POS = range(len(infile))
    START_POS = range(3, 83, 4)
    RESULT_LIST = []
    for position in TOT_POS:
        if position%10000 == 0:
            print(position, end=" ")
        non_zero_indices = []
        for start in START_POS:
            non_zero_indices_temp = list(np.nonzero(list(infile.iloc[position, start:start+4]))[0])
            non_zero_indices.extend(non_zero_indices_temp)
        non_zero_indices = list(np.unique(non_zero_indices))
        result = []
        for start in START_POS:
            result.append(list(infile.iloc[position, start:start+4])[non_zero_indices[0]])
            result.append(list(infile.iloc[position, start:start+4])[non_zero_indices[1]])
        RESULT_LIST.append(result)
    return pd.DataFrame(RESULT_LIST, columns=colnames)

# ==============================

column_names = ['fFRA_1_1', 'fFRA_1_2', 'fGRE_12_1', 'fGRE_12_2', 'wFRA_1_1', 'wFRA_1_2',
                'fSPA_3_1', 'fSPA_3_2', 'fSPA_2_1', 'fSPA_2_2', 'fGRE_10_1', 'fGRE_10_2',
                'wGRE_6_1', 'wGRE_6_2', 'fCRO_5_1', 'fCRO_5_2', 'fGRE_11_1', 'fGRE_11_2',
                'wITA_4_1', 'wITA_4_2', 'wGRE_7_1', 'wGRE_7_2', 'wTUR_9_1', 'wTUR_9_2',
                'fCRO_6_1', 'fCRO_6_2', 'fCYP_13_1', 'fCYP_13_2', 'fGRE_9_1', 'fGRE_9_2',
                'fGRE_8_1', 'fGRE_8_2', 'wSPA_2_1', 'wSPA_2_1', 'wCRO_5_1', 'wCRO_5_2',
                'fGRE_7_1', 'fGRE_7_2', 'wTUR_10_1', 'wTUR_10_2']


Dlabrax_maf_baypass = BayPass_genotyping_file(biallelic_Dlabrax_maf, column_names)
```
